Для определения высокоррелированных атрибутов построим матрицу корреляции, а для ее визуализации - тепловую карту. На тепловой карте более яркие цвета обозначают высокоррелированные параметры, менее яркие - низкоколеррированные. Используем функции библиотек seaborn и matplotlib.

```python
sns.heatmap(df_new.corr(numeric_only = True), annot=True, cmap='coolwarm',linewidths=0.5, linecolor='black' )
plt.title('Матрица корреляции')
plt.show()
```

Проанализируем тепловую карту.

Касаемо оценки тесноты связи, в разных источниках встречаются разные классификации (градации).
Например, используем эту:

от ±0,81 до ±1,0 – сильная сила связи;

от ±0,61 до ±0,8 – умеренная сила связи;

от ±0,41 до ±0,6 – слабая сила связи;

от ±0,21 до ±0,4 – очень слабая сила связи;

от 0 до ±0,2 – связь отсутствует.

Тогда:

Высокая положительная корреляция - avg_income и education_expenditure

Высокая отрицательная корреляция - IQ и rank

Умеренная положительная корреляция - rank и avg_temp

Умеренная отрицательная корреляция - education_expenditure и rank, avg_income и rank, IQ и avg_temp

Слабая положительная корреляция - education_expenditure и IQ, avg_income и IQ

Слабая отрицательная корреляция - education_expenditure и avg_temp, avg_income и avg_temp

Пар атрибутов без связи нет, но самая низкая корреляция у атрибутов avg_temp и avg_income.

Вывод:

Проанализируем полученные диаграммы рассеивания.

На диагонали матрицы находятся гистограммы распределения, по которым мы можем оценить форму распределения:

rank - все столбцы примерно одной высоты, поэтому у этой гистограммы равномерное распределение, то есть все значения имеют примерно одинаковую вероятность появления

IQ и temp – у этих гистограмм мы наблюдаем следующее соотношение характеристик распределения: среднее значение<медиана<мода, что говорит о гистограмме, скошенной влево, или об отрицательной ассиметрии. Следовательно в этих данных преобладают относительно высокие значения, а более низкие значения менее представлены. Это может указывать на наличие аномальных низких значений (выбросов) , что мы и наблюдали в “ящике с усами” IQ, или на неравномерное распределение.

education_expenditure и avg_income - у этих гистограмм мы наблюдаем следующее соотношение характеристик распределения: среднее значение>медиана>мода, что говорит о гистограмме, скошенной вправо, или об положительной ассиметрии. Следовательно в этих данных преобладают относительно низкие значения, а более высокие значения менее представлены. Это может указывать на наличие аномальных высоких значений (выбросов) , что мы и наблюдали в “ящиках с усами” этих атрибутов.


Графики, расположенные вне диагонали - диаграммы рассеивания для каждой пары атрибутов. По ним можно судить о взаимосвязи между атрибутами.

1) IQ + rank : почти прямая убывающая линия говорит о высокой отрицательной корреляции между атрибутами. Об этом же говорит значение в матрице корреляции, равное -0.97. Это означает, что они сильно влияют друг на друга, а именно с уменьшением одного параметра, возрастает другой. Действительно, чем выше показатель IQ, тем меньше значение ранга страны (именно меньше число, отвечающее за ранг страны. Разумеется, это означает, что страна находится выше в рейтинге, как бы на 1,2,3 и тд месте)

2) У следующих пар атрибутов диаграммы рассеивания напоминают график экспоненты (в классическом виде, а так же с по разному расставленными знаками у экспоненты и x). Следовательно это экспоненциальные диаграммы рассеяния, то есть точки образуют кривую, напоминающую экспоненту, что может указывать на то, что зависимость между атрибутами подчиняется экспоненциальному закону:

education_expenditure + rank ; education_expenditure + IQ ; avg_income + IQ ;
 avg_income + rank ; avg_temp + rank ; avg_temp + IQ .

Так же расположение точек соответствует ранее сделанным выводам о умеренной и слабой корреляции, а направление экспоненты – отрицательной и положительной корреляции соответственно (возрастающая функция соответствует положительной корреляции, а отрицательная - убывающей) . И еще мы наблюдаем особые точки, которые отдалены от остального кластера точек – это выбросы.

 3) avg_income + education_expenditure : очень напоминает прямую возрастающую линию, что соответствует высокой положительной корреляции. Однако наблюдаются выбросы, которые “портят картину”

 4) avg_temp + education_expenditure : одна из двух пар со слабой корреляцией, потому набор точек рассеян. Наблюдаются выбросы.

5) avg_temp + avg_income : самый “рассеянный” набор точек. Соответствует самой слабой корреляцию, по сравнению с другими. Наблюдаются выбросы.
